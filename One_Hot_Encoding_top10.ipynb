# One Hot Encoding

```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('mercedesbenz.csv', usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
# to See all columns

pd.set_option("display.max_columns", None)
```

```python
data.head()
data.shape
```

```python
# Lets have unique category in each column

for cols in data.columns:
    print(cols,':', len(data[cols].unique()), 'labels'  )
```

```python
# Lets find out how many more columns we will obtain after Hot Encoding
pd.get_dummies(data,drop_first=True).shape

# So , Originally 6 columns will be increased to 117 columns
```

```python
## KDD 2009 cup:  One Hot Encoding with 10 most frequent labels

## Lets find top 10 variables in a column X1 which have 27 labels

data.X1.value_counts(ascending=False).head(10)

# Put top 10 variables in a list
top_10=[x for x in data.X1.value_counts().head(10).index]

# Create new columns with these top 10 variables

for i in top_10:
    data[i]= np.where(data['X1']==i,1,0)
    
data[['X1']+top_10  ].head(10)
```

```python
## Now do this for all 6 columns

def one_hot_top10(df,column,top_10   ):
    for i in top_10:
        df[column+"_"+i ]= np.where(data[column]==i,1,0)
    
    
#Now find top 10 variables from each column and call function    
```

```python
# For X2  because X1 already has been done directly

top_10 =[x for x in data.X2.value_counts().head(10).index  ]

one_hot_top10(data,'X2',top_10)
```

```python
# For X3:

top_10=[x for x in data.X3.value_counts().head(10).index]
top_10

one_hot_top10(data, 'X3', top_10)
data
```
