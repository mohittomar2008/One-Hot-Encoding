# Label Encoding

```python
import pandas as pd
import datetime
```

```python
# we are taking last 20days date then acording to days, we will label encode these
```

```python
now=datetime.datetime.now()
now
```

```python
date_list=[now- datetime.timedelta(days=x) for x in range(0,20) ]
```

```python
date_list
```

```python
# convert this last 2o days into dataframe
df= pd.DataFrame(date_list)
```

```python
# set column name to Day
df.columns=['Day']
```

```python
# extract weekdays name

df['day_of_week']= df['Day'].dt.day_name()
```

```python
df
```

```python
# Engineer categorical variable by ordinal number replacement
weekday_map= {'Monday':1,"Tuesday":2,"Wednesday":3,"Thursday":4,"Friday":5,"Saturday":6,"Sunday":7}
```

```python
df['day_ordinal']= df['day_ofweek'].map(weekday_map)
```
